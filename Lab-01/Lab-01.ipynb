```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("bank.csv", sep = ';')
df.head()
```

```python
df.tail()
```

```python
def replace_marital(val):
  if val == 'single':
    return 0
  else:
    return 1

df['marital'] = df['marital'].apply(replace_marital)
df.head()
```

```python
df["housing"] = df["housing"].map({"yes": 1, "no": 0}.get)
df.head()
```

```python
df["loan"] = df["loan"].replace({"yes": 1, "no": 0})
df.head()
```

```python
df["loan"] = df["loan"].replace({
"yes": 1,
"no": 0})
df.head()
```

```python
df["job"].unique()
```

```python
df["job"] = df["job"].replace({'unemployed': 0,
                   'services': 0,
                   'management': 1,
                   'blue-collar': 0,
                   'self-employed': 0,
                   'technician': 1,
                   'entrepreneur': 1,
                   'admin.': 0,
                   'student': 1,
                   'housemaid': 0,
                   'retired': 0,
                   'unknown': np.nan})
df.head()
```

```python
df["month"].unique()
```

```python
df.month = df.month.map({
    'oct': 10,
    'may': 5,
    'apr': 4, 
    'jun': 6,
    'feb': 2,
    'aug': 8,
    'jan': 1,
    'jul': 7,
    'nov': 11,
    'sep': 9,
    'mar': 3,
    'dec': 12
})
df.head(10)
```

```python
df["education"].unique()
```

```python
df.education = df.education.map({
    'primary': 1, 
    'secondary': 2,
    'tertiary': 3,
    'unknown': np.nan
})
```

```python
df.head(10)
```

```python
df["poutcome"].unique()
```

```python
df.poutcome = df.poutcome.map({
    'unknown': np.nan, 
    'failure': 1,
    'other': 2,
    'success': 3
})
df.head(10)
```

```python
df["balance"] = df["balance"].apply(lambda v: (v - df["balance"].min())/ (df["balance"].max() - df["balance"].min()))
```

```python
df.head(10)
```

```python
df["pdays"] = df["pdays"].apply(lambda v: (v - df["pdays"].min())/ (df["pdays"].max() - df["pdays"].min()))
```

```python
df.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
df["duration"] = scaler.fit_transform(df[["duration"]])
df["pdays"] = scaler.fit_transform(df[["pdays"]])
df.head()
```
