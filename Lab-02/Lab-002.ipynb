# Data Preprocessing

```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("DT.csv")
df.head()
```

```python
df.tail()
```

```python
def replace_PlayTennis(val):
    if val == 'No':
        return 0
    else:
        return 1
    
df['PlayTennis'] = df['PlayTennis'].apply(replace_PlayTennis)
df.head(14)
```

```python
df["Wind"] = df["Wind"].map({"strong": 1, "weak": 0}.get)
df.head(14)
```

```python
df["Humidity"] = df["Humidity"].map({"high": 1, "normal": 0}.get)
df.head(14)
```

```python
df["Temp"].unique()
```

```python
df["Temp"] = df["Temp"].replace({
    'cold': 0,
    'mild': 1,
    'hot': 2,
    'unknown': np.nan})

df.head(14)
```

```python
df["Outlook"].unique()
```

```python
df["Outlook"] = df["Outlook"].replace({
    'rain': 0,
    'overcast': 1,
    'sunny': 2,
    'unknown': np.nan})

df.head(14)
```

```python
df["Day"].unique()
```

```python
df["Day"] = df["Day"].replace({
    'D1': 1,
    'D2': 2,
    'D3': 3,
    'D4': 4,
    'D5': 5,
    'D6': 6,
    'D7': 7,
    'D8,': 8,
    'D9': 9,
    'D10': 10,
    'D11': 11,
    'D12': 12,
    'D13': 13,
    'D14': 14,
    'D15': 15,
    'unknown': np.nan})

df.head(14)
```

```python
df.drop(columns=['Unnamed: 6'], inplace=True)
```

```python
df.head(14)
```
